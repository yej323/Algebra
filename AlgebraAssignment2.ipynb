Vertex stabiliser subgroups are isomorphic to cyclic groups of order 3; I will call this Z3.

Stabiliser subgroups of pairs of opposite verteces are C3xZ2, equivalently the dihedral group of order 3.

Stabiliser subgroups of edges are Z2

Stabiliser subgroups pairs of edges are Z2xZ2

Face stabiliser subgroups are Z5

stabiliser subgroups of pairs of faces are Z5xZ2 or the dihedral group of order 5

In the list above, G1 will be the first nontrivial group, G2 the second, etc.
